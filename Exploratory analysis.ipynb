### Moving average function

```python
def mov_avg(df,window):
    mean = df.rolling(window).mean()
    ma = mean
    return ma 
```

### CREATING DUMMY TIME SERIES

```python
# VOLUME SHOCKS
infosys_df['Volume_shock']=((abs(infosys_df.Volume-infosys_df.Volume.shift())/infosys_df.Volume)>=0.10).astype(int)
infosys_df.head()
```

```python
# PRICE SHOCKS
infosys_df['Price_shock']=(((abs(infosys_df.Close-infosys_df['Prev Close']))/infosys_df.Close)>=0.02).astype(int)
infosys_df.head()
```

```python
# price shock without volume shock
infosys_df['priceshock_wihout_volshock']=((infosys_df.Price_shock==1) & (infosys_df.Volume_shock==0)).astype(int)
infosys_df.head()
```

## Handling unequal time series

```python
idx = pd.date_range('1-1-2015','30-12-2016')
idx
```

```python
aug_infydf = infosys_df.reindex(idx)
print(infosys_df.shape)
aug_infydf.isnull().sum()
```

#### As we can see there are 235 holidays in between 2015 to 2016 out of total 730 days i.e. around 32% of observations is missing. Hence, filling out these values might not be a good strategy and we might end up adding our own biases to the dataset.
##### We will do feature engineering by adding a new indicator variable named 'holiday' which will tell our model about the missing data due to holidays.

```python
# adding new feature which indicates if a date is holiday
aug_infydf['Holiday']=aug_infydf['Close'].isnull().astype(int)
```

```python
# as sklearn cannot handle Nan, hence we will fill missing values of holidays with 0
aug_infydf.fillna(0,inplace=True)
aug_infydf.head()
```
